# **Import Libraries**

```python
import numpy as np
import cv2 as cv
from skimage.feature import peak_local_max
```

```python
# Define the file paths
path_jpg = '/content/img_roi.jpg'
path_png = '/content/road.png'

# Read the images
# cv2.imread() loads an image directly into a NumPy array
edges = cv.imread(path_jpg, 0)
img = cv.imread(path_png)


# edges = cv.cvtColor(edges, cv.COLOR_BGR2RGB)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
rho_step = 9
theta_step = 0.261

def hough_transform(edges: np.ndarray):
  img_height = edges.shape[0]
  img_width  = edges.shape[1]

  max_rho = np.sqrt(img_width ** 2 + img_height ** 2)

  thetas = np.arange(0, np.pi, theta_step)
  rhos = np.arange(-max_rho, max_rho, rho_step)

  accumulator_width  = len(thetas)
  accumulator_height = len(rhos)

  accumulator = np.zeros([accumulator_height, accumulator_width])

  sins  = np.sin(thetas)
  coss = np.cos(thetas)

  xs, ys = np.where(edges > 0)

  for x, y in zip(xs, ys):
    for theta_index in range(len(thetas)):
      rho = x * coss[theta_index] + y * sins[theta_index]
      rho_index = np.argmin(np.abs(rho - rhos))

      accumulator[rho_index, theta_index] += 1

  local_max_indices = peak_local_max(accumulator, threshold_rel = 0.5)
  final_rhos_indices, final_thetas_indices =  map(list, zip(*local_max_indices))
  final_rhos = rhos[final_rhos_indices]
  final_thetas = thetas[final_thetas_indices]

  return accumulator, np.vstack([final_rhos, final_thetas]).T
```

```python
def draw_lines(image: np.ndarray, hough_lines: np.ndarray):
    image_with_lines = np.copy(image)
    line_mask = np.zeros(image.shape[:2], dtype=np.uint8)

    for rho, theta in hough_lines:
        point = polar_to_cartesian(rho, theta)
        direction = np.array([point[1], -point[0]])
        point1 = np.round(point + 1000 * direction).astype(int)
        point2 = np.round(point - 1000 * direction).astype(int)
        line_mask = cv.line(img=line_mask, pt1=point1, pt2=point2, color=255, thickness=1)

    line_pixels = line_mask != 0
    min_line_width = np.inf
    intersection_row = 0

    for row_idx in range(line_pixels.shape[0]):
        row_line_pixels = line_pixels[row_idx]
        col_indices = np.argwhere(row_line_pixels)
        if len(col_indices) > 1:
            line_width = col_indices[-1] - col_indices[0]
            if line_width < min_line_width:
                min_line_width = line_width
                intersection_row = row_idx

    lower_mask = np.zeros_like(line_mask)
    lower_mask[intersection_row:] = 1
    final_mask = (line_pixels * lower_mask).astype(bool)

    image_with_lines[final_mask] = np.array([0, 0, 255])

    return image_with_lines

def polar_to_cartesian(rho: np.ndarray, angle: np.ndarray):
    return rho * np.array([np.sin(angle), np.cos(angle)])
```

```python
_,lines = hough_transform(edges)
# Show the image with the lines found
lines_img = draw_lines(img, lines)

win_name = "hough"
cv.namedWindow(win_name)
cv.imshow(win_name, lines_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
